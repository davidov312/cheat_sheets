### Object oriented example

```python
class Employee:
    'Common base class for all employees'
    empCount = 0

    def __init__(self, name, salary):
        self.name = name
        self.salary = salary
        Employee.empCount += 1
   

    def displayCount(self):
        print("Total Employee %d" % Employee.empCount)
   
    
    def displayEmployee(self):
        print("Name : ", self.name,  ", Salary: ", self.salary)
               

emp1 = Employee("Zara", 2000)
emp2 = Employee("Manni", 5000)

emp1.displayEmployee()
emp2.displayEmployee()
print("Total Employee %d" % Employee.empCount)
```

## Pandas group by in parallel

```python
import pandas as pd
from multiprocessing import get_context, cpu_count

def my_func(df):
    df_out = df.copy()
    df_out['square'] = df_out.my_val.pow(2)
    return df_out

df = pd.DataFrame(data={'my_val': range(15),
                        'my_group': list(range(3))*5})


multi_proc_flag = True

if not multi_proc_flag: #single core processing
    result_list = []
    for _, group_df in df.groupby('my_group'):
        result_list.append(my_func(group_df))
else: # multi core processing
    pool = get_context().Pool(processes=cpu_count())
    results = [pool.apply_async(my_func, args=(group_df))
                for _, group_df in df.groupby('my_group')]
    
    pool.close()
    pool.join()
    result_list = [res.get() for res in results if res.ready()]
result_df = pd.concat(result_list, axis=0, ignore_index=True, sort=False)  
result_df = result_df.sort_values(by='my_val')
result_df
```

```python
import pandas as pd
from multiprocessing import get_context, cpu_count

def my_func(n):
    return n+10

multi_proc_flag = True

if not multi_proc_flag: #single core processing
    result_list = []
    for i in range(10):
        result_list.append(my_func(i))
else: # multi core processing
    pool = get_context().Pool(processes=cpu_count())
    results = [pool.apply_async(my_func, args=(i))
                for i in range(10)]
    
    pool.close()
    pool.join()
    result_list = [res.get() for res in results if res.ready()]
result_list
```


# Function attributes

```python
import sys


def my_add(a, b):
    return a + b


def my_sub(a, b):
    return a - b


for func_str in ['my_add', 'my_sub']:
    func = getattr(sys.modules[__name__], func_str)
    print(func(2,3))
```
